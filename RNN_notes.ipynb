### Procesado de datos

Como una red neuronal no sólo puede procesar números lo mejor es convertir las palabras a un número, a este proceso se le llama tokenización.
El problema es que si se le asigna números inevitablemente unos serán mucho mayores que otros haciendo que la red interprete que tienen un mayor valor semántico. La solución es tokenizar en base a vectores usando one-hot encoding

En el que que las clases serán cada una de las palabras del vocabulario por lo que tenemos un vocabulario con 10000 palabras tendríamos vectores de longitud 10000 lleno de ceros con un 1 en la poción de la palabra

En un espacio geometrico cada palabra tendría la misma distancia que todas las demas pero a nivel semántico sabemos que algunas palaras están más cecanas unas a otras. Otro problema es que al ser vectores muy grandes que en su mayoría tiene ceros realmenete ocupa mucho espacio inecesario.

#### Embedding
La idea del compactamiento se basa en ordenar los datos dados ciertos atributos, una forma de hacer esto es pasarle estos vectores enormes a una red neuronal para que aprenda los atributos semanticos de cada palabra mediante un entrenamiento, de esta manera dada una palabra la red neuronal retornará un vector que define su posición semantica en un espacio geometrico.
Usualmente se usa aprendizaje no supervizado para que la red encuentre estros patrones por si sola

Lo bueno es que existen modelos de embedding pre entrenados que cumplen esta tarea de forma generica como Word2vec

### Exploding gradients y Vanishing gradients


Hablamos de exploding gradients cuando el algoritmo asigna una importancia exageradamente alta a los pesos sin mucha razón, y esto genera un problema en el entrenamiento, en este caso el problema se puede resolver si se truncan o reducen los gradientes.

Hablamos de Vanishing Gradients cuando los valores de un gradiente son demasiado pequeños y el modelo deja de aprender o requiere demasiado tiempo, este problema e puede resolver con las capas LSTM (long short term memory)

Las LSTM son una extención de las RNN, una LSTM puede leer, escribir y borrar información de su memoria y aprende con el tiempo que información es importante y cual no.
Las puertas de una LSTM son ánalogas a una forma sigmoide lo que permite incorporarlas al proceso de back propagation.
